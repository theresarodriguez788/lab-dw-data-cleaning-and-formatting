## Exercise 1: Cleaning Column Names

To ensure consistency and ease of use, standardize the column names of the dataframe. Start by taking a first look at the dataframe and identifying any column names that need to be modified. Use appropriate naming conventions and make sure that column names are descriptive and informative.

*Hint*:

- *White spaces in column names should be replaced by `_`*

```python
import pandas as pd

# Load the data
url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv'
data_bc = pd.read_csv(url)

# Print the original column names
print("Original column names:")
print(data_bc.columns)

# Create a dictionary for replacements
replacement_dict = {
    "Customer": "customer",
    "ST": "state",
    "GENDER": "gender",
    "Education": "education",
    "Customer Lifetime Value": "customer_lifetime_value",
    "Income":"income",
    "Monthly Premium Auto": "monthly_premium_auto",
    "Number of Open Complaints":"number_of_open_complaints",
    "Policy Type": "policy_type",
    "Vehicle Class": "vehicle_class",
    "Total Claim Amount": "total_claim_amount",
    
}

# Rename the columns using the replacement dictionary
data_bc.rename(columns=replacement_dict, inplace=True)

# Print the first few rows of the updated dataframe
print("Updated dataframe with standardized column names:")
data_bc
```

## Exercise 2: Cleaning invalid Values

The dataset contains columns with inconsistent and incorrect values that could affect the accuracy of our analysis. Therefore, we need to clean these columns to ensure that they only contain valid data.

Note that this exercise will focus only on cleaning inconsistent values and will not involve handling null values (NaN or None).

*Hint*:
- *Gender column contains various inconsistent values such as "F", "M", "Femal", "Male", "female", which need to be standardized, for example, to "M" and "F".*
- *State abbreviations be can replaced with its full name, for example "AZ": "Arizona", "Cali": "California", "WA": "Washington"*
- *In education, "Bachelors" could be replaced by "Bachelor"*
- *In Customer Lifetime Value, delete the `%` character*
- *In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"*

```python
# Step 1: Standardize the values in the "Gender" column
data_bc['gender'] = data_bc['gender'].str.lower().replace({'femal': 'f', 'male': 'm'})

# Step 2: Replace state abbreviations with full names in the "State" column
state_mapping = {
    'AZ': 'Arizona',
    'Cali': 'California',
    'WA': 'Washington'
}
data_bc['state'] = data_bc['state'].replace(state_mapping)

# Step 3: Standardize education levels in the "Education" column
data_bc['education'] = data_bc['education'].replace({'Bachelors': 'Bachelor'})

# Step 4: Remove the "%" character from values in the "Customer Lifetime Value" column
data_bc['customer_lifetime_value'] = data_bc['customer_lifetime_value'].str.replace('%', '')

# Step 5: Replace certain vehicle classes with a more general category in the "Vehicle Class" column
vehicle_mapping = {
    'Sports Car': 'Luxury',
    'Luxury SUV': 'Luxury',
    'Luxury Car': 'Luxury'
}
data_bc['vehicle_class'] = data_bc['vehicle_class'].replace(vehicle_mapping)

# Print the first few rows of the cleaned dataframe to verify the changes
print("Cleaned dataframe:")
data_bc
```

## Exercise 3: Formatting data types

The data types of many columns in the dataset appear to be incorrect. This could impact the accuracy of our analysis. To ensure accurate analysis, we need to correct the data types of these columns. Please update the data types of the columns as appropriate.

It is important to note that this exercise does not involve handling null values (NaN or None).

*Hint*:
- *Customer lifetime value should be numeric*
- *Number of open complaints has an incorrect format. Look at the different values it takes with `unique()` and take the middle value. As an example, 1/5/00 should be 5. Number of open complaints is a string - remember you can use `split()` to deal with it and take the number you need. Finally, since it should be numeric, cast the column to be in its proper type.*

```python
import pandas as pd

# Ensure "Customer Lifetime Value" is numeric
data_bc['customer_lifetime_value'] = pd.to_numeric(data_bc['customer_lifetime_value'], errors='coerce')

# Step 2: Correct the format of "Number of Open Complaints" column and convert it to numeric type
# Convert to string first
data_bc['number_of_open_complaints'] = data_bc['number_of_open_complaints'].astype(str)

# Split the string and take the second part
data_bc['number_of_open_complaints'] = data_bc['number_of_open_complaints'].str.split('/').str[1]

# Convert to numeric type
data_bc['number_of_open_complaints'] = pd.to_numeric(data_bc['number_of_open_complaints'], errors='coerce')

# Print the data types of all columns to verify the changes
print("Updated data types:")
print(data_bc.dtypes)

data_bc
```

## Exercise 4: Dealing with Null values

Identify any columns with null or missing values. Identify how many null values each column has. You can use the `isnull()` function in pandas to find columns with null values.

Decide on a strategy for handling the null values. There are several options, including:

- Drop the rows or columns with null values
- Fill the null values with a specific value (such as the column mean or median for numerical variables, and mode for categorical variables)
- Fill the null values with the previous or next value in the column
- Fill the null values based on a more complex algorithm or model (note: we haven't covered this yet)

Implement your chosen strategy to handle the null values. You can use the `fillna()` function in pandas to fill null values or `dropna()` function to drop null values.

Verify that your strategy has successfully handled the null values. You can use the `isnull()` function again to check if there are still null values in the dataset.

Remember to document your process and explain your reasoning for choosing a particular strategy for handling null values.

After formatting data types, as a last step, convert all the numeric variables to integers.

```python
# Step 1: Identify columns with null values and determine the number of null values in each column
null_counts = data_bc.isnull().sum()
print("Columns with null values and their respective counts:")
print(null_counts)

# Step 2: Decide on a strategy for handling null values
# Let's fill null values with the mean for numeric columns and the mode for categorical columns

# Step 3: Implement the chosen strategy
# Fill null values with the mean for numeric columns
numeric_columns = data_bc.select_dtypes(include=['int64', 'float64']).columns
data_bc[numeric_columns] = data_bc[numeric_columns].fillna(data_bc[numeric_columns].mean())

# Fill null values with the mode for categorical columns
categorical_columns = data_bc.select_dtypes(include=['object']).columns
data_bc[categorical_columns] = data_bc[categorical_columns].fillna(data_bc[categorical_columns].mode().iloc[0])

# Step 4: Verify that null values have been handled
null_counts_after = data_bc.isnull().sum()
print("\nColumns with null values after handling:")
print(null_counts_after)

# Step 5: Convert all numeric variables to integers
data_bc[numeric_columns] = data_bc[numeric_columns].astype(int)

# Print the updated data types to confirm conversion
print("\nUpdated data types:")
print(data_bc.dtypes)
```

# Bonus: Challenge 2: creating functions on a separate `py` file

Put all the data cleaning and formatting steps into functions, and create a main function that performs all the cleaning and formatting.

Write these functions in separate .py file(s). By putting these steps into functions, we can make the code more modular and easier to maintain.

*Hint: autoreload module is a utility module in Python that allows you to automatically reload modules in the current session when changes are made to the source code. This can be useful in situations where you are actively developing code and want to see the effects of changes you make without having to constantly restart the Python interpreter or Jupyter Notebook kernel.*

```python
# cleaning_functions.py

import pandas as pd

def standardize_column_names(df):
    """
    Standardize column names by converting them to lowercase and replacing white spaces with underscores.
    """
    df.columns = df.columns.str.lower().str.replace(' ', '_')
    return df

def clean_gender_column(df):
    """
    Clean the 'gender' column by standardizing the values.
    """
    df['gender'] = df['gender'].str.lower().replace({'femal': 'f', 'female': 'f', 'male': 'm', 'm': 'm'})
    return df

def clean_state_column(df):
    """
    Clean the 'state' column by replacing state abbreviations with full names.
    """
    state_mapping = {
        'AZ': 'Arizona',
        'Cali': 'California',
        'WA': 'Washington'
    }
    df['state'] = df['state'].replace(state_mapping)
    return df

def clean_education_column(df):
    """
    Standardize education levels in the 'education' column.
    """
    df['education'] = df['education'].replace({'Bachelors': 'Bachelor'})
    return df

def clean_customer_lifetime_value(df):
    """
    Remove the '%' character from values in the 'customer_lifetime_value' column and convert to numeric.
    """
    df['customer_lifetime_value'] = df['customer_lifetime_value'].str.replace('%', '').astype(float)
    return df

def clean_vehicle_class_column(df):
    """
    Replace certain vehicle classes with a more general category in the 'vehicle_class' column.
    """
    vehicle_mapping = {
        'Sports Car': 'Luxury',
        'Luxury SUV': 'Luxury',
        'Luxury Car': 'Luxury'
    }
    df['vehicle_class'] = df['vehicle_class'].replace(vehicle_mapping)
    return df

def correct_data_types(df):
    """
    Correct the data types of various columns.
    """
    df['customer_lifetime_value'] = pd.to_numeric(df['customer_lifetime_value'], errors='coerce')
    df['number_of_open_complaints'] = df['number_of_open_complaints'].str.split('/').str[1]
    df['number_of_open_complaints'] = pd.to_numeric(df['number_of_open_complaints'], errors='coerce')
    return df

def handle_missing_values(df):
    """
    Handle missing values by filling numeric columns with mean and categorical columns with mode.
    """
    numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns
    df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean())
    
    categorical_columns = df.select_dtypes(include=['object']).columns
    df[categorical_columns] = df[categorical_columns].fillna(df[categorical_columns].mode().iloc[0])
    
    return df

def drop_duplicates_and_reset_index(df):
    """
    Drop duplicate rows and reset the index.
    """
    df = df.drop_duplicates().reset_index(drop=True)
    return df

data_bc
```

# Bonus: Challenge 3: Analyzing Clean and Formated Data

You have been tasked with analyzing the data to identify potential areas for improving customer retention and profitability. Your goal is to identify customers with a high policy claim amount and a low customer lifetime value.

In the Pandas Lab, we only looked at high policy claim amounts because we couldn't look into low customer lifetime values. If we had tried to work with that column, we wouldn't have been able to because customer lifetime value wasn't clean and in its proper format. So after cleaning and formatting the data, let's get some more interesting insights!

Instructions:

- Review the statistics again for total claim amount and customer lifetime value to gain an understanding of the data.
- To identify potential areas for improving customer retention and profitability, we want to focus on customers with a high policy claim amount and a low customer lifetime value. Consider customers with a high policy claim amount to be those in the top 25% of the total claim amount, and clients with a low customer lifetime value to be those in the bottom 25% of the customer lifetime value. Create a pandas DataFrame object that contains information about customers with a policy claim amount greater than the 75th percentile and a customer lifetime value in the bottom 25th percentile.
- Use DataFrame methods to calculate summary statistics about the high policy claim amount and low customer lifetime value data. To do so, select both columns of the dataframe simultaneously and pass it to the `.describe()` method. This will give you descriptive statistics, such as mean, median, standard deviation, minimum and maximum values for both columns at the same time, allowing you to compare and analyze their characteristics.

```python
import pandas as pd

# Load the cleaned data
cleaned_file_path = 'cleaned_data.csv'
data_bc = pd.read_csv(cleaned_file_path)

# Review the statistics for total_claim_amount and customer_lifetime_value
print("Statistics for Total Claim Amount:")
print(data_bc['total_claim_amount'].describe())

print("\nStatistics for Customer Lifetime Value:")
print(data_bc['customer_lifetime_value'].describe())

# Calculate the 75th percentile for total_claim_amount and the 25th percentile for customer_lifetime_value
total_claim_amount_75th = data_bc['total_claim_amount'].quantile(0.75)
customer_lifetime_value_25th = data_bc['customer_lifetime_value'].quantile(0.25)

# Identify customers with high policy claim amounts and low customer lifetime value
high_claim_low_lifetime_customers = data_bc[
    (data_bc['total_claim_amount'] > total_claim_amount_75th) &
    (data_bc['customer_lifetime_value'] < customer_lifetime_value_25th)
]

# Calculate summary statistics for the high policy claim amount and low customer lifetime value data
summary_stats = high_claim_low_lifetime_customers[['total_claim_amount', 'customer_lifetime_value']].describe()

print("\nSummary statistics for customers with high policy claim amounts and low customer lifetime value:")
print(summary_stats)
```
